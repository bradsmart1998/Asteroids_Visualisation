```python
# Import dependencies
import pandas as pd
import requests
import json
```

```python
# Import the api_key 
from config import api_key
```

```python
# Create the URL 
url = "https://api.nasa.gov/neo/rest/v1/feed?"
start_date = "2023-01-01"
end_date = "2023-01-08"
```

```python
# Assemble the query URL
query_url = f"{url}start_date={start_date}&end_date={end_date}&api_key={api_key}"
```

```python
# Get the response
response = requests.get(query_url).json()
```

```python
# Create a list to store the asteriod data
asteriods_list = []
asteriods = response["near_earth_objects"]
# Loop through the response and pull the data for each asteriod
for asteriod in asteriods:
    ast = response["near_earth_objects"][asteriod]
    asteriods_list.append(ast)
```

```python
# import itertools and get seperate the list within a list
import itertools
asteriods__flist = list(itertools.chain(*asteriods_list))
```

```python
# Put the data into a dataframe
asteriod_df = pd.DataFrame(asteriods__flist)
asteriod_df
target_cols = ["id", "name", "absolute_magnitude_h", "estimated_diameter",
               "is_potentially_hazardous_asteroid", "close_approach_data"]
asteriod_df = asteriod_df[target_cols]
asteriod_df.head()
```

```python
# Pull the data from the JSON within the "estimated_diameter" column and append the data to a list
km_min_list = []
km_max_list = []
ft_min_list = []
ft_max_list = []
for i in range(0,len(asteriod_df)):
    km_min = asteriod_df.iloc[i,3]['kilometers']["estimated_diameter_min"]
    km_min_list.append(km_min)
    km_max = asteriod_df.iloc[i,3]['kilometers']["estimated_diameter_max"]
    km_max_list.append(km_max)
    ft_min = asteriod_df.iloc[i,3]['feet']["estimated_diameter_min"]
    ft_min_list.append(ft_min)
    ft_max = asteriod_df.iloc[i,3]['feet']["estimated_diameter_max"]
    ft_max_list.append(ft_max)
```

```python
# Pull the data from the JSON within the "close_approach_data" column and append the data to a list
kph_list = []
mph_list = []
miss_dist_list_km = []
miss_dist_miles_list = []
for i in range(0,len(asteriod_df)):
    kph = asteriod_df.iloc[i,5][0]['relative_velocity']['kilometers_per_hour']
    kph_list.append(kph)
    mph = asteriod_df.iloc[i,5][0]['relative_velocity']['miles_per_hour']
    mph_list.append(mph)
    miss_km = asteriod_df.iloc[i,5][0]['miss_distance']['kilometers']
    miss_dist_list_km.append(miss_km)
    miss_miles = asteriod_df.iloc[i,5][0]['miss_distance']['miles']
    miss_dist_miles_list.append(miss_miles)
```

```python
# Drop any N/A values
asteriod_df.dropna(inplace=True)
# Add the columns to the dataframe
asteriod_df['km_min'] = km_min_list
asteriod_df['km_max'] = km_max_list
asteriod_df['ft_min'] = ft_min_list
asteriod_df['ft_max'] = ft_max_list
asteriod_df['velocity_kph'] = kph_list
asteriod_df['velocity_mph'] = mph_list
asteriod_df['miss_distance_km'] = miss_dist_list_km
asteriod_df['miss_distance_miles'] = miss_dist_miles_list
```

```python
# Create the final dataframe including the new columns
target_cols = ["id", "name", "absolute_magnitude_h", "is_potentially_hazardous_asteroid",
               'km_min','km_max', 'ft_min', 'ft_max', 'velocity_kph', 'velocity_mph',
               'miss_distance_km','miss_distance_miles']
asteriod_df = asteriod_df[target_cols]
# Rename two columns
asteriod_df.rename(columns={"absolute_magnitude_h": "magnitude",
                            "is_potentially_hazardous_asteroid": "hazardous"}, inplace=True)
# Final df
asteriod_df.head()
```

```python
# Info to convert the columns
asteriod_df.info()
```

```python
# Change to numeric values
asteriod_df.id = pd.to_numeric(asteriod_df.id)
asteriod_df.velocity_kph = pd.to_numeric(asteriod_df.velocity_kph)
asteriod_df.velocity_mph = pd.to_numeric(asteriod_df.velocity_mph)
asteriod_df.miss_distance_km = pd.to_numeric(asteriod_df.miss_distance_km)
asteriod_df.miss_distance_miles = pd.to_numeric(asteriod_df.miss_distance_miles)
# New Info
asteriod_df.info()
```
